```python
import pandas as pd
import numpy as np



def inverse(pri):
    global change
    change = [0]
    for i in range (1, len(pri)):
        change.append(pri[i]-pri[i-1])
    inv = [pri[0]]
    for i in range (1,len(change)+1):
        inv.append((inv[i-1]-change[i-1])%12)
    
    return inv[1:]

def newrow(i):
    row = [i]
    for i in range (1, len(change)+1):
        row.append((row[i-1]+change[i-1])%12)
    return row[1:]


o_prime = [11,10,2,3,7,6,8,4,5,0,1,9]
o_inverse = inverse(o_prime)

df = pd.DataFrame(columns = o_prime, index = o_inverse)


for i in o_prime:
    df.loc[i] = newrow(i)

df
```
